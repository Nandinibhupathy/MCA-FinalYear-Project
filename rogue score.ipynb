```python
from evaluate import load
# Load the ROUGE metric
import evaluate
rouge = evaluate.load('rouge')
candidates = ["What do seed plants called angiosperms produce? : seeds",
              "What do enzymes help convert? : Food",
              "What is the Sun's innermost layer? : core",
              "What is the term for the process of sexual reproduction involving haploid gametes and producing a diploid zygote through fertilization? : fertilization",
              "What do sepals protect? : flower"
             
             ]

references = ["Where do angiosperms produce seeds in flowers? : ovaries",
             "Most of the chemical reactions in the body are facilitated by what? : enzymes",
             "What is the suns innermost layer called? : core",
             "Sexual reproduction involves haploid gametes and produces a diploid zygote through what process? : fertilization",
             "What protects a developing flower while it is still a bud? : sepals"]
results=[]

for i in range(5):
    results.append(rouge.compute(predictions=candidates, references=references))
print(results)
```
